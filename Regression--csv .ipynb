```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Dell\Downloads\kc_house_data.csv',delimiter = ',')
df
```

```python
df.set_index('id',inplace=True)
df
```

```python
x= df.floors
y = df.price
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(x,y,marker='*',markerfacecolor='red')
plt.xlabel('No. of floors',labelpad = 20,fontsize = 'medium')
plt.ylabel('Price in thousand $',labelpad = 20,fontsize='medium')
plt.title('Price  v/s  No.of Floors',pad=20,fontsize='xx-large')
plt.xlim(left = 1, right = 5)
plt.show()
```

```python
x= df.floors
y = df.bedrooms
```

```python
plt.plot(x,y,marker='*',markerfacecolor='red',markersize=15)
plt.xlabel('No. of floors',labelpad = 20,fontsize = 'medium')
plt.ylabel('No.of bedrooms',labelpad = 20,fontsize='medium')
plt.title('No.of bedrooms  v/s  No.of floors',pad=20,fontsize='xx-large')
plt.ylim(bottom = 1 , top = 5)
plt.show()
```

```python
x_axis = df.floors
```

```python
plt.hist(x_axis,bins=4,edgecolor='red')
plt.xlabel('No. of floors',labelpad = 20,fontsize = 'medium')
plt.ylabel('No.of houses',labelpad = 20,fontsize='medium')
plt.title('frequency  v/s  No.of floors',pad=20,fontsize='xx-large')
plt.show()
```

```python
x_axis = df.floors
y_axis = df.sqft_living
```

```python
plt.bar(x_axis,y_axis,width=0.3,align = 'center')
plt.xlabel('No. of floors',labelpad = 20,fontsize = 'medium')
plt.ylabel('sqft of living area',labelpad = 20,fontsize='medium')
plt.title('sqft  v/s  No.of floors',pad=20,fontsize='xx-large')
plt.xlim(left = 1, right = 4)
plt.show()
```

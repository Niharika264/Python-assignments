```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Dell\Downloads\lockdownindiawarningzones.csv')
df.set_index('S.no',inplace=True)
df
```

```python
x = df.Zone
```

```python
plt.hist(x,bins=10,align='left')
plt.xlabel('Contamination Zone',labelpad = 20,fontsize = 'medium')
plt.ylabel('Frequency of Zones',labelpad = 20,fontsize='medium')
plt.title('covid-19 india contamination zones',pad=20,fontsize='xx-large')
plt.show()
```

```python
x1 = df.Zone
y1 = df.State
```

```python
fig = plt.figure(figsize=(15,15),facecolor='green')
plt.plot(x1,y1,marker='o',markerfacecolor='g')
plt.xlabel('States',labelpad = 20,fontsize = 'x-large')
plt.ylabel('Contamination zones',labelpad = 20,fontsize='x-large')
plt.title('covid-19 india contamination zones',pad=20,fontsize='xx-large')
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8,13),facecolor='orange')
plt.scatter(x1,y1,marker='o')
plt.xlabel('Contamination Zone',labelpad = 20,fontsize = 'medium')
plt.ylabel('Frequency of Zones',labelpad = 20,fontsize='medium')
plt.title('covid-19 india contamination zones',pad=20,fontsize='xx-large')
plt.show()
```

```python
## Plotting for first 20 rows to avoid clumpsyness
data = df.head(20)
x_data = data.Zone
y_data = data.District
```

```python
fig = plt.figure(figsize(9,13))
plt.plot(x_data,y_data,marker='o',markerfacecolor='g')
plt.xlabel('States',labelpad = 20,fontsize = 'x-large')
plt.ylabel('Contamination zones',labelpad = 20,fontsize='x-large')
plt.title('covid-19 india contamination zones',pad=20,fontsize='xx-large')
plt.show()
```
